```python
source = 'https://data.culture.gouv.fr/api/explore/v2.1/catalog/datasets/aides-a-la-presse-classement-des-titres-de-presse-aides/exports/csv?lang=fr&timezone=Europe%2FBerlin&use_labels=true&delimiter=%3B'
```

```python
aides = (
    pd
    .read_csv(source, sep=';')
)

aides
```

```python
aides_annee = (
    aides
    .pivot_table(
        index='Bénéficiaires',
        columns = 'Année',
        values = 'Aides directes (1) (en €)'
    )
    .assign(
        total = lambda df: df.sum(axis=1)
    )
)

(
    aides_annee
    .sort_values('total', ascending = False)
    .head(50)
    .style
    .format(thousands = " ", na_rep='', precision = 0)

)
```

```python
aides_annee.sort_values('total', ascending = False).to_csv('datasets/aides-annees.csv')
```

```python
aides_annee_total = (
    aides_annee
    .sum()
    .to_frame()
    .T
)

aides_annee_total.pipe(format_table)
```

```python
sheet_url = 'https://docs.google.com/spreadsheets/d/18NX2VrdvC0O8sSXTDsSLlaiFhRd6RQ3_4MUE9MBrb9U/gviz/tq?tqx=out:csv&sheet=Sheet1'

#pd.read_csv(sheet_url)

aides_annee_clean = (
    aides_annee
    #.query('total > 100000')
    .join(pd.read_csv(sheet_url).set_index('Bénéficiaires'))
    .assign(
        recodage = lambda df: df.recodage.str.upper()
    )
    .groupby('recodage')
    .sum()
    .sort_values('total', ascending=False)
)

aides_annee_clean.pipe(format_table)
```

```python
aides_annee_clean.to_csv('datasets/aides-total_sup_100000-clean.csv')
```
