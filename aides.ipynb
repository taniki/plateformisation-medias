```python
import pandas as pd
```

```python
source = 'https://data.culture.gouv.fr/api/explore/v2.1/catalog/datasets/aides-a-la-presse-classement-des-titres-de-presse-aides/exports/csv?lang=fr&timezone=Europe%2FBerlin&use_labels=true&delimiter=%3B'
```

```python
aides = (
    pd
    .read_csv(source, sep=';')
)

aides
```

```python
(
    aides
    .pivot_table(
        index='Bénéficiaires',
        columns = 'Année',
        values = 'Aides directes (1) (en €)'
    )
    .assign(
        total = lambda df: df.sum(axis=1)
    )
    .sort_values('total', ascending = False)
    .head(50)
    .style
    .format(thousands = " ", na_rep='', precision = 0)
)
```
