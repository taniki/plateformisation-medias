```python
# https://docs.google.com/spreadsheets/d/1zmdPsrOou_hFo8g7Zf-3jC6sZ-omVE_j69MCP3L1dDg/edit?usp=sharing

entreprises = (
    pd
    .read_csv('https://docs.google.com/spreadsheets/d/1zmdPsrOou_hFo8g7Zf-3jC6sZ-omVE_j69MCP3L1dDg/gviz/tq?tqx=out:csv&sheet=entreprises')
    .set_index('siren')
)

entreprises
```

```python
comptes = (
    pd
    .read_csv('data/ratios_inpi_bce.csv', sep=";")
    .query('type_bilan == "C"')
    .assign(
        date_cloture_exercice = lambda df: pd.to_datetime(df.date_cloture_exercice),
        cloture_exercice_annee = lambda df: pd.to_datetime(df.date_cloture_exercice).dt.to_period('y')
    )
)

comptes
```

```python
(
    entreprises
    [['nom']]
    .join(
        entreprises
        .join(comptes.set_index('siren'))
        .pivot_table(
            index = 'siren',
            columns = 'cloture_exercice_annee',
            values = 'chiffre_d_affaires',
            aggfunc= 'last'
        )
    )
    .loc[entreprises.index]
    #.set_index('nom')
    .style
    .format(thousands = " ", precision = 0)
)
```

```python
(
    entreprises
    .join(comptes.set_index('siren'))
    .pivot_table(
        index = 'nom',
        columns = 'cloture_exercice_annee',
        values = 'chiffre_d_affaires'
    )
    .sum()
    .to_frame()
    .T
    .style
    .format(thousands = " ", precision = 0)
)
```

```python
(
    entreprises
    .join(comptes.set_index('siren'))
    .pivot_table(
        index = 'nom',
        columns = 'cloture_exercice_annee',
        values = 'resultat_net'
    )
    .sum()
    .to_frame()
    .T
    .style
    .format(thousands = " ", precision = 0)
)
```
